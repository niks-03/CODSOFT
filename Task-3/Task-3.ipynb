```python
sales_data = pd.read_csv("advertising.csv")
```

```python
fig,axes=plt.subplots(1,3,figsize=(23,6))
sbn.barplot(x=sales_data["TV"], y=sales_data["Sales"], palette="cool",ax=axes[0])
sbn.barplot(x=sales_data["Radio"], y=sales_data["Sales"], palette="cool",ax=axes[1])
sbn.barplot(x=sales_data["Newspaper"], y=sales_data["Sales"], palette="cool",ax=axes[2])
```

```python
fig, axes=plt.subplots(1,3,figsize=(23,6))
sbn.scatterplot(x="TV", y="Sales", data=sales_data, ax=axes[0])
sbn.scatterplot(x="Radio", y="Sales", data=sales_data, ax=axes[1])
sbn.scatterplot(x="Newspaper", y="Sales", data=sales_data, ax=axes[2])
```

```python
x = sales_data.drop(columns=["Sales"])
y = sales_data["Sales"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
train_prediction = model.predict(x_train)
train_prediction
```

```python
model.score(x_train,y_train)
```

## Predicted values

```python
test_prediction = model.predict(x_test)
test_prediction
```

#### Score of model

```python
model.score(x_test, y_test)
```
